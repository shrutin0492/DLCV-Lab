```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load the image
image = cv2.imread('/content/noisy_image.jpg', 0) #Add image path here

# Convert the noisy image to 8-bit unsigned integer format
noisy_image = np.uint8(image)

# Add some noise to the image
row, col = image.shape
gauss = np.random.normal(0, 1, (row, col))
noisy_image = image + gauss * 50

# Ensure the noisy image is in uint8 format
noisy_image = np.uint8(noisy_image)

# Applying Gaussian blur
gaussian_blur = cv2.GaussianBlur(noisy_image, (5, 5), 0)

# Applying Median blur
# Ensure that the image passed to medianBlur is in uint8 format
median_blur = cv2.medianBlur(np.uint8(noisy_image), 5)

# Display results
plt.subplot(231), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(232), plt.imshow(noisy_image, cmap='gray'), plt.title('Noisy Image')
plt.subplot(233), plt.imshow(gaussian_blur, cmap='gray'), plt.title('Gaussian Blur')
plt.subplot(234), plt.imshow(median_blur, cmap='gray'), plt.title('Median Blur')
plt.show()
```
